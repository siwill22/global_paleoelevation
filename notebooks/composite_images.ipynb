```python
import numpy as np
from PIL import Image
from IPython.display import display
import moviepy.editor as mpy
```

```python
def get_concat_h_blank(im1, im2, color=(0, 0, 0)):
    dst = Image.new('RGB', (im1.width + im2.width, max(im1.height, im2.height)), color)
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst

def get_concat_v_blank(im1, im2, color=(255, 255, 255), x_shift=0):
    dst = Image.new('RGB', (max(im1.width, im2.width), im1.height + im2.height), color)
    dst.paste(im1, (0, 0))
    dst.paste(im2, (x_shift, im1.height))
    return dst


    
```

```python
def two_panels(model_name,calibration,mohometer_description_string, space_bin_size, age1, age2):
    
    im1 = Image.open('../images/sequence_{:s}/elevations_{:s}_{:s}_{:d}_{:0.0f}Ma.png'.format(model_name,
                                                                                         calibration,
                                                                                         mohometer_description_string,
                                                                                         space_bin_size,
                                                                                         age1))
    
    im2 = Image.open('../images/sequence_{:s}/elevations_{:s}_{:s}_{:d}_{:0.0f}Ma.png'.format(model_name,
                                                                                                  calibration,
                                                                                                  mohometer_description_string,
                                                                                                  space_bin_size,
                                                                                                  age2))
    
    im3 = get_concat_h_blank(im1.resize((int(im1.width * im2.height/im1.height), im2.height)), im2)
    
    return im3


def two_panels_residuals(model_name,calibration,mohometer_description_string, space_bin_size, age1, age2):
    
    im1 = Image.open('../images/sequence_{:s}/elevation_residuals_{:s}_{:s}_{:d}_{:0.0f}Ma.png'.format(model_name,
                                                                                         calibration,
                                                                                         mohometer_description_string,
                                                                                         space_bin_size,
                                                                                         age1))
    
    im2 = Image.open('../images/sequence_{:s}/elevation_residuals_{:s}_{:s}_{:d}_{:0.0f}Ma.png'.format(model_name,
                                                                                                  calibration,
                                                                                                  mohometer_description_string,
                                                                                                  space_bin_size,
                                                                                                  age2))
    
    im3 = get_concat_h_blank(im1.resize((int(im1.width * im2.height/im1.height), im2.height)), im2)
    
    return im3
    
    

 
    
model_name='Paleomap'
calibration = 'luffi'
mohometer_description_string = '41'
space_bin_size = 5
time_step = 10


age_mins = [0,100,200,300,400]

for age_min in age_mins:
    
    age_max = age_min+80
    
    im_list = []
    im_list_residuals = []
    for left_age in np.arange(age_min, age_max+time_step, time_step*2):
        im_list.append(two_panels(model_name, calibration, mohometer_description_string, space_bin_size, left_age, left_age+10))
        im_list_residuals.append(two_panels_residuals(model_name, calibration, mohometer_description_string, space_bin_size, left_age, left_age+10))


    final_im = im_list[0]
    for im in im_list[1:]:
        final_im = get_concat_v_blank(final_im,im)

    final_im.save('../images/composite_images/sequence_{:s}_{:s}_{:s}_{:d}_{:d}Ma.png'.format(model_name,
                                                                                              calibration,
                                                                                              mohometer_description_string,
                                                                                              space_bin_size,
                                                                                              age_min))

    final_im = im_list_residuals[0]
    for im in im_list_residuals[1:]:
        final_im = get_concat_v_blank(final_im,im)

    final_im.save('../images/composite_images/sequence_residuals_{:s}_{:s}_{:s}_{:d}_{:d}Ma.png'.format(model_name,
                                                                                                        calibration,
                                                                                                        mohometer_description_string,
                                                                                                        space_bin_size,
                                                                                                        age_min))
    
```
